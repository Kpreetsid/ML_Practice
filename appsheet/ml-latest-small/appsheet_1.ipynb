```python
#What are the 25 most rated movies by total voting size?

most_rated = movie_rating.groupby('title').size().sort_values(ascending=False)[:25]
most_rated
```

```python
#What are the 25 most rated movies when considering the mean of the rating?

movie_stats = movie_rating.groupby('title').agg({'rating': [np.size, np.mean]})
atleast_100 = movie_stats['rating']['size'] >= 100
movie_stats[atleast_100].sort_values([('rating', 'mean')], ascending=False)[:25]
```

```python
#Grouping the above query by genre

movie_stats = movie_rating.groupby(['genres', 'title']).agg({'rating': [np.size, np.mean]})
atleast_100 = movie_stats['rating']['size'] >= 100
movie_stats[atleast_100].sort_values([('rating', 'mean')], ascending=False)[:50]
```
