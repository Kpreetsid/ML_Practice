```python
def get_user_similar_movies( user1, user2 ):
  common_movies = ratings[ratings.userId == user1].merge(ratings[ratings.userId == user2], on = "movieId", how = "inner" )

  return common_movies.merge(movies , on = 'movieId' )
```

```python
get_user_similar_movies( 310, 247 )
```

```python
rating_mat = ratings.pivot( index='movieId', columns='userId', values = "rating" ).reset_index(drop=True)
```

```python
rating_mat.fillna( 0, inplace = True )
```

```python
rating_mat.shape
```

```python
rating_mat.head(10)
```

```python
movie_sim = 1 - pairwise_distances(rating_mat.as_matrix(), metric="correlation")
```

```python
movie_sim.shape
```

```python
movie_sim_df = pd.DataFrame(movie_sim)
```

```python
movie_sim_df.head(10)
```

```python
movies['similarity'] = movie_sim_df.iloc[0]
movies.columns = ['movieId', 'title', 'genres', 'similarity']
```

```python
movies.head()
```

```python
movies.sort_values( ["similarity"], ascending = False )[1:10]
```

```python
def get_similar_movies( movieId, topN = 5 ):
  movies['similarity'] = movie_sim_df.iloc[movieId -1]
  top_n = movies.sort_values( ["similarity"], ascending = False )[0:topN]
  print( "Similar Movies to: ", )
  return top_n
```

```python
get_similar_movies( 118 )
```

```python
get_similar_movies( 127, 10 )
```
