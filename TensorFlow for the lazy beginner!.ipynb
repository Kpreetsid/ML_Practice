```python
import matplotlib.pyplot as plt

year = [1990, 1992, 1994, 1996, 1998]
pop = [ 1.42, 1.58, 1.67, 1.79, 1.98]

year2 = [2000, 2002, 2004, 2006, 2008]
pop2 = [ 2.42, 2.58, 2.67, 2.79, 2.92]

plt.plot(year, pop, label='First Decade')
plt.plot(year2, pop2, label='Next Decade')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population\nProjections')
plt.legend()
plt.show()
```

```python
plt.scatter(year, pop)
plt.scatter(year2, pop2)
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```
